```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.Connection('Chinook_Sqlite.sqlite')
cursor = conn.cursor()
```

```python
cursor.execute('''
SELECT t.Name, SUM(i.Quantity) AS Sold, DENSE_RANK() OVER(ORDER BY SUM(i.Quantity) DESC)
FROM InvoiceLine i
JOIN Track t    
ON t.TrackId = i.TrackId
GROUP BY i.TrackId 
ORDER BY Sold DESC
               ''')
cols = cursor.fetchall()
df = pd.DataFrame(cols, columns=['Name', 'Sold', 'TOP PRODUCTS'])
df.to_csv("top_products.csv", sep=',')
df
```

```python
cursor.execute('''
SELECT c.Country, SUM(i.Total), ROW_NUMBER() OVER(ORDER BY SUM(i.Total) DESC) AS RANK
FROM Invoice i 
JOIN Customer c
ON c.CustomerId = i.CustomerId
GROUP BY c.Country
ORDER BY SUM(i.Total) DESC
             ''')
data = cursor.fetchall()
df2 = pd.DataFrame(data, columns=['Country', 'Total_Revenue', 'RANK'])
df2.to_csv('revenue_per_country.csv')
df2
```

```python
cursor.execute(''' 
SELECT STRFTIME('%Y-%m', InvoiceDate) AS MONTH, SUM(Total)
FROM Invoice 
GROUP BY MONTH
ORDER BY MONTH
''')
data = cursor.fetchall()
df3 = pd.DataFrame(data, columns=['MONTH', 'Monthly_Performance'])
df3.to_csv("monthly_performance.csv")
df3
```
